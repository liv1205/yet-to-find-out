```python
# the silly little questions off my silly little journal

date = datetime.today().strftime("%d/%m/%Y")
print(date)
print("---------------------")
print("First Questions")
print("---------------------")


def get_ans_quali(quali):
    while True:
        try:
          ans_quali = input(f"{quali}")
          if ans_quali != "°":
            return ans_quali.lower()
            break
          else:
              print("wtf?")
        except ValueError:
            print("Please type something valid")


quali: str = {
      "name":"Type your name: ",
      "goodie":"Write a good thing that happened today: ",
      "bad":"If something that bothered you happened today, write down what it was: ",
      "lon":"How well do you think you have avoided isolation yourself? "
}

ans_quali: str= {}


for code, quali in quali.items():
    ans_quali[code] = get_ans_quali(quali)

df = pd.DataFrame([ans_quali])
print("----------------------")
time.sleep(0.5)
print(df)

# agee

age: int = int(input("How old are you? "))


#numbers next


def get_answer(question):
    while True:
        try:
            answer = int(input(f"{question} (0–5): "))
            if 0 <= answer <= 5:
                return answer
            else:
                print("Please enter a number between 0 and 5.")
        except ValueError:
            print("Invalid input. Please enter a number.")



# Make sure this dictionary is wrapped in curly braces and has commas between entries
questions = {
    "wellness": "How did you feel today? (0–5)",
    "rl_partner": "How do you see your current love life? (0–5)",
    "scl_anxt": "How well did you handle social interactions today? (0–5)",
    "rl_family": "How was your relationship with your family today? (0–5)",
    "sl_sleep": "How well did you sleep last night? (0–5)",
    "pd_productivity": "How productive did you feel today? (0–5)",
    "em_mood": "How was your mood overall today? (0–5)"
}



answers = {}

for code, question in questions.items():
    answers[code] = get_answer(question)
```

# storaaageee

```python
# Project WAAITT (we're all in this together)


waaitt = {**{"date": date}, **ans_quali, **{"age": age}, **answers}

df = pd.DataFrame([waaitt],columns=[
      "date",
      "name",
      "age",
      "goodie",
      "bad",
      "lon",
      "wellness",
      "rl_partner",
      "rl_family",
      "scl_anxt",
      "pd_productivity",
      "em_mood"
])


print( )
print( )
print( "  Results " )
print("-------------------------")
print(df)


df.to_excel("new_entry.xlsx", index=False)
print(os.getcwd())
```
